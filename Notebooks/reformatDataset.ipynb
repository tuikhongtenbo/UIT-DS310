```python
import json
from collections import defaultdict
```

```python
input_file = '../dataset/bge_reranker_train_triplets.json'
output_file = '../dataset/train_reranker_main.json'

with open(input_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

query_dict = defaultdict(lambda: {'pos': set(), 'neg': set()})
```

```python
for item in data:
    query = item['query']
    positive = item['positive']
    negative = item['negative']
    query_dict[query]['pos'].add(positive)
    query_dict[query]['neg'].add(negative)

reformatted_data = []

for query, values in query_dict.items():
    reformatted_data.append({
        'query': query,
        'pos': list(values['pos']),
        'neg': list(values['neg'])
    })

with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(reformatted_data, f, ensure_ascii=False, indent=2)
```

```python
for i, sample in enumerate(reformatted_data[:2]):
    print(f"\nSample {i+1}:")
    print(f"  Query: {sample['query'][:100]}...")
    print(f"  Số positives: {len(sample['pos'])}")
    print(f"  Số negatives: {len(sample['neg'])}")
```
